# Seattle Airbnb Content-Based Recommendation System
- Dataset Source: https://www.kaggle.com/airbnb/seattle
- Reasons for choosing this dataset:
    - It is a very good dataset for an content-based recommendation because the presence of numerous text columns which describes the Airbnb listing in every area. E.g. "description", "neighborhood", "transit", "host_about"
    - This really suits the context of a Chatbot well because basically, users can just tell us anything about their ideal airbnb and we can ask them very broad questions
    - Will be able to create a very natural conversation experience, unlike mechanical conversations where Chatbot ask specific questions from user and expect specific answers
    - There are also other useful columns such as "review score" which can be really helpful in recommending an Airbnb
    - Furthermore, the dataset comes with a Review dataset which contains 80000+ reviews for all the listings. This opens up other possibilities to our recommendation system such as Sentiment analysis and Topic identification to provide a better user experience with additional features
<hr>

## Plan
- Using chatbot to first ask for filter columns e.g. budget (price), property_type, room_type)
    - Filter columns will be used to filter out AirBnBs that does not meet the hard requirement
- Then ask what are users' expectations of the AirBnb they want to stay in
    - E.g. host personality, ambience, interior design, location, neighbourhood, ...
- Concatenate the potential description columns to produce a document for each Airbnb listing
- Natural Language processing of the document of each Airbnb
- Based on user's description, insert it into the Airbnb listing dataframe, act as if it is a listing itself
- Generate the TFIDF vector for the whole Airbnb listing dataframe
- Compute the cosine similarity of user description's vector array with all the other listing vector arrays
- Recommend the top 2 similar AirBnb listing to user
<hr>
- Additional Features:
    - Sentiment Analysis for the reviews dataset to generate a sentiment score for each Airbnb listing
    - Topic Analysis - Identify top 3 bigrams in recommended listings' reviews and output to user with the bigrams in original sample review

## Term Frequency-Inverse Document Frequency (TF-IDF) Vectors
- TF-IDF vectors reflect a word's importance to an document, adjusting its importance by taking into account the number of documents containing that word
    - The importance of a word = TF * IDF
- Term Frequency (TF) is the relative frequency of a word in a document
    - Calculated as (The Word's instances)/(Total Number of Words)
- Inverse Document Frequence is the relative count of documents containing that word
    - Calculated as log(Number of documents/Number of document with the word)
- When a word appears more time in a document -> TF increases -> TF * IDF will increase -> more important
- When a word appears in many documents -> (Number of documents/Number of document with the word) will decrease -> IDF decreases -> TF * IDF decreases -> less important 
<hr>
- The sklearn.feature_extraction.text.TfidfVectorizer will compute the TF-IDF vectors for the Airbnb listing dataset 
    - TfidfVectorizer will output an csr_matrix; similar to numpy array but does not record words that does not appear in its document
    - Each row in the csr_matrix is the document of each Airbnb listing (potentional desc columns concatenated)
    - Each column is the word that appears in the document with its importance (TF * IDF)

## Cosine-similarity
- After generating the TF-IDF vectors of every Airbnb Listing, this is equvalient to representing each listing in a multi-dimensional space with each dimension being a word. 
- In this multi-dimensional space, each listing is at different point, determinied by its words' TF * IDF values 
    - This is also known as word vectors
- The similarity of two listings' vectors can be computed by consine-similarity; cos(c) = v1*v2 / v1.v2
    - c is actually the angle between the 2 vectors. 
    - cos(c) = 1 -> Vector in same direction -> identical
    - cos(c) = 0 -> Vector perpendicular -> very different
- We will use linear_kernel to compute the similarity between listings' vectors

````python
# Generate the cosine similarity matrix for each listing against every other listing
cosine_sim = linear_kernel(listing_matrix, listing_matrix)
````

## Addition Feature: Topic Analysis - Identify Keywords in Listing's Reviews
- Although the host_url will be provided to user, but instead of the user reading through thousands of reviews, we can possibly provide keywords from the reviews to allow the user to quickly identify the main topics | features | highlights of the reviews of the recommended listings
- We will be using Collections to identify the top 10 bigrams from recommended listing's reviews 
    - Collection is expressions of multiple words which commonly co-occur
    - Bigram - 2 words that co-occur frequenty
<hr>
- Reference: 
    - http://www.nltk.org/howto/collocations.html

````python
# Import the processed review dataset from "Review Sentiment Analysis" notebook
# It has NaN filled and non-english comments removed. So no need to process here again
reviews = pd.read_csv("reviews_processed.csv")
````

````python
# Filter out reviews for the recommended listings
reviews = reviews[(reviews['listing_id'] == recommendation['id'].values[0]) | (reviews['listing_id'] == recommendation['id'].values[1])]
print("Number of reviews for recommended listings: {}".format(reviews.shape[0]))

# Group the reviews by their listing_id and join all comments of each listing into a doc
listing_reviews = reviews.groupby(['listing_id'])['comments'].apply(lambda x: " ".join(x))
````

````python
# Tokenize the doc for each listing
def tokenize(doc):
    tokens = word_tokenize(doc)
    # Remove punctuations
    tokens = [token for token in tokens if token not in string.punctuation]
    # word_tokenize will tokenize contractions. E.g. "didn't" -> ["did", "n't"], "Sally's" -> ["Sally", "'s"]  
    # Do not want that to happen as they will always co-occur
    # Will affect the accuracy of top bigrams identified
    result = []
    for i, token in enumerate(tokens):
        if token[0] == "'" or token == "n't":
            result[-1] = result[-1] + token
        else:
            result.append(token)
    return(result)

listing_reviews = listing_reviews.map(tokenize)
listing_reviews
````

````python
# Identify the top 3 bigrams for each listing
def bigramify(doc):
    bigram_measures = BigramAssocMeasures()
    finder = BigramCollocationFinder.from_words(doc)
    # If the bigram didn't appear more than 3 times, we will not consider them
    finder.apply_freq_filter(3)
    # Return the top 3 bigrams
    return finder.nbest(bigram_measures.pmi, 3)

listing_bigrams = listing_reviews.map(bigramify)
listing_bigrams
````

````python
# Return the top bigrams with the reviews that they appear in 
# Easier for user to understand the context
def sample_reviews(listing_id):
    # Retrieve the bigrams for the recommended listing
    bigrams = listing_bigrams[listing_id]
    # Retrieve all the reviews for the recommended listing
    review_texts = reviews[reviews['listing_id'] == listing_id]['comments'].values
    sample_reviews = []
    for bigram in bigrams:
        # Filter out the reviews that contain the bigrams
        comb_bigram = " ".join(bigram)
        potential_reviews = list(filter(lambda review : comb_bigram in review, review_texts))
        # Randomly chose a review that contain the bigram to be displayed
        sample_review = random.choice(potential_reviews)
        # Remove some unwanted characters
        sample_review = sample_review.replace("\n","").replace("\r","")
        # Display the review with bigrams highlighted
        sample_review = sample_review.replace(comb_bigram, "****" + comb_bigram + "****")
        # Display some characters before and after the bigram
        index = sample_review.index(comb_bigram)
        sample_text = "..." + sample_review[index-50 : index+50] + "..."
        sample_reviews.append(sample_text)
    return sample_reviews

sample_reviews(recommendation['id'].values[0])
````
